```python
numbers = [1, 2, 2, 3, 3, 3, 4]
frequency = {}
for num in numbers:
    if num in frequency:
        frequency[num] += 1
    else:
        frequency[num] = 1
print(frequency)
```

```python
li=[45, 78, 12, 89, 56, 90, 67]
for i in range(len(li)):
    for j in range(i+1, len(li)):
        if li[i] < li[j]:
            li[i], li[j] = li[j], li[i]
print(li[:3])
```
