```python
# Dataset (rows = students, cols = subjects)
marks = np.array([
    [90, 85, 88],
    [78, 82, 80],
    [92, 89, 94],
    [60, 70, 65]
])
# Average per student
avg = np.mean(marks, axis=1)
# Highest marks
highest = np.max(marks)
# Pass/Fail (pass if avg >= 75)
result = avg >= 75
print("Average:", np.round(avg,2))
print("Highest:", highest)
print("Pass/Fail:", result)
```
