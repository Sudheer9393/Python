```python
s="Python"
count=0
for i in s:
    count+=1
    print(i,end="")
print()
print(count)
```

```python
s1="Python"
s=s1[::-1]
print(s)
rev=""
for i in s1:
    rev=i+rev
print(rev)
```

```python
s="Data Science"
vowels="aeiouAEIOU"
count=0
for i in s:
    if i in vowels:
        count+=1
        print(i,end=" ")
print()
print(count)
```

```python
s= "PyThOn"
up_count=0
lo_count=0
for i in s:
    if i.isupper():
        up_count+=1
    else:
        lo_count+=1
print(up_count)
print(lo_count)
        
```

```python
s=input()
s1=s[::-1]
if s==s1:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
s="programming"
dup=""
for i in s:
    if i not in dup:
        dup+=i
print(dup)
```

```python
s="programming"
freq={}
for i in s:
    if i not in freq:
        freq[i]=1
    else:
        freq[i]+=1
print(freq)
```
