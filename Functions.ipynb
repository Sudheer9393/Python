```python
def occurances(data):
    count=0
    for i in data:
        if i>50:
            count+=1
    return count
data = [20,55,70,30,90]
print(occurances(data))
```

```python
def smaller(data):
    small=data[0]
    for i in data:
        if i<small:
            small=i
    return small
data = [5,9,2,7]
print(smaller(data))
```

```python
def even(data):
    li=[]
    for i in data:
        if i%2==0:
            li.append(i)
    return li
data=[1,2,3,4,5,6]
print(even(data))
```

```python
def vowel(s):
    count=0
    v="aeiooAEIOU"
    for i in s:
        if i in v:
            count+=1
    return count
s="DataScience"
print(vowel(s))
```

```python
def minandmax(n):
    return max(n),min(n)
data = [5,2,9,1]
minandmax(data)
    
```

```python
def normal(data):
    no=[]
    for i in data:
        normalized = (i- min(data)) / (max(data) - min(data))
        no.append(normalized)
    return no
data = [10,20,30]
print(normal(data))
```

```python
def nor(data):
    return [(i-min(data))/(max(data)-min(data)) for i in data]
data=[10,20,30]
print(nor(data))
```

```python
def occ(data):
    di={}
    for i in data:
        if i not in di:
            di[i]=1
        else:
            di[i]+=1
    return di
data=[1,2,2,3,3,3]
occ(data)
```

```python
def duplicates(data):
    du=[]
    for i in data:
        if data.count(i)>1 and i not in du:
            du.append(i)
    return du
data=[1,2,3,2,4,1]
duplicates(data)
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(9)) 
```

```python
def factorial(n):
    if n == 0 or n == 1: 
        return 1
    else:
        return n * factorial(n-1) 
print(factorial(5)) 
```
